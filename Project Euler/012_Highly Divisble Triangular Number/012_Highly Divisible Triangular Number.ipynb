#Highly Divisible Triangular Number

The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

 1: 1
 3: 1,3
 6: 1,2,3,6
10: 1,2,5,10
15: 1,3,5,15
21: 1,3,7,21
28: 1,2,4,7,14,28
We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

```python
def triangle_divisors(limit):
    
    triangle_number = 0
    a = 1
    
    while True:
        
        triangle_number += a
        num_range = list(range(1,triangle_number))
        #divisors = {i for i in num_range if triangle_number % i == 0}
        divisors = 0
        for i in num_range:
            if triangle_number % i == 0:
                divisors += 1
        #divisors = list(filter(lambda x:triangle_number % x == 0, num_range))
        if divisors >= limit:
            return divisors,triangle_number
        else:
            a += 1
```

```python
triangle_divisors(500)
```
